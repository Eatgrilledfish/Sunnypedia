```python
import pandas as pd
cancer_mortality_data = pd.read_csv("./cancerMortalityRates_utf8/Table S2a.1-Table 1.csv", engine='python', skiprows=5,skipfooter=19)
print(cancer_mortality_data.head())
```

```python
# Create a boolean mask to check where the 'Count' column has the value '.'
dot_mask = (cancer_mortality_data['Count'] == '.')

# Filter the DataFrame based on the boolean mask
entries_with_dot = cancer_mortality_data[dot_mask]

number_of_entries_with_dot = dot_mask.sum()

print("Number of entries where 'Count' column has the value '.':", number_of_entries_with_dot)
```

```python
cancer_mortality_data_cleaned = cancer_mortality_data[['Data type', 'Cancer group/site', 'Year', 'Sex','Age group (years)','Count','Age-specific rate\n(per 100,000)']]
# Printing the cleaned dataset
print(cancer_mortality_data_cleaned.head())
```

```python
cancer_mortality_data_cleaned_1 = cancer_mortality_data_cleaned[(cancer_mortality_data_cleaned['Cancer group/site'] == 'Melanoma of the skin') | (cancer_mortality_data_cleaned['Cancer group/site'] == 'Non-melanoma skin cancer (rare types)')]
print(cancer_mortality_data_cleaned )
cancer_mortality_data_cleaned=cancer_mortality_data_cleaned_1[cancer_mortality_data_cleaned_1['Age group (years)']=='All ages combined']
```

```python
#filter all ages combined
#filter all the actual data
```

```python
cancer_mortality_data_cleaned ['Count'] = cancer_mortality_data_cleaned ['Count'].astype(str).str.replace(',', '').astype(int)
```

```python
skin_cancer_mortality_data_sum = cancer_mortality_data_cleaned.groupby('Year')['Count'].sum().reset_index(name ='Total mortality')
print(skin_cancer_mortality_data_sum)
```
